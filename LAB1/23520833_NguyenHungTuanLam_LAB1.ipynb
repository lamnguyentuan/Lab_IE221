```python
t=979
sol=[((t-7*y-9*z)//2,y,z)
     for z in range(1,t//9+1)
     for y in range(1,(t-9*z)//7+1)
     if (t-7*y-9*z)>0 and (t-7*y-9*z)%2==0]
print(f"Kết quả các nghiệm (x, y, z) là:  {sol}")
```

Bài 50

```python
import random
print("Nhập số lượng vé mua: ")
number_of_ticket = int(input())
listOfTicket = [
    random.sample(range(1,46),6) for i in range(number_of_ticket)
]
print(f"Danh sách vé: {listOfTicket}")
lotteryNumber = random.sample(range(1,46),6)
print(f"Kết quả số trúng: {lotteryNumber}")
calculate_prize = lambda ticket: (
        10000000 if len(set(ticket).intersection(lotteryNumber)) == 6 else
    10000000 if len(set(ticket).intersection(lotteryNumber)) == 5 else
    300000 if len(set(ticket).intersection(lotteryNumber)) == 4 else
    30000 if len(set(ticket).intersection(lotteryNumber)) == 3 else
    0
)


total_prize = sum(map(calculate_prize, listOfTicket))

for i, ticket in enumerate(listOfTicket, 1):
    prize = calculate_prize(ticket)
    print(f"Vé {i} - Số: {ticket} - Tiền thưởng: {prize} đồng")


print(f"Tổng tiền thưởng: {total_prize} đồng")
```
